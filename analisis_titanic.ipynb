```python
# Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Tampilkan semua kolom & baris
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)

# Load Data
df = pd.read_excel('titanic.xlsx')  # Ganti path jika perlu
```

```python
# 1. Observasi Awal
print("Head:")
print(df.head())

print("\nTail:")
print(df.tail())

print("\nSample:")
print(df.sample(5))

print("\nInfo:")
df.info()
```

```python
# 2. Statistical Summary
print("\nStatistical Summary:")
print(df.describe(include='all'))
```

```python
# 3. Cek Duplikasi
duplikat = df.duplicated().sum()
print(f"\nJumlah data duplikat: {duplikat}")

# Hapus duplikat jika ada
df = df.drop_duplicates()
```

```python
# 4. Missing Values
missing = df.isnull().sum()
missing_percent = (missing / len(df)) * 100

missing_summary = pd.DataFrame({
    'Missing Values': missing,
    'Percentage (%)': missing_percent
})
print("\nMissing Value Summary:")
print(missing_summary)

# Handling missing value (contoh pada kolom age)
if 'age' in df.columns:
    mean_age = df['age'].mean()
    df['age'].fillna(mean_age, inplace=True)
    print("\nMissing value pada kolom 'age' sudah diisi dengan rata-rata.")
```
